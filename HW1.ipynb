```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def original(input_image):
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(input_image); plt.title("Original");
    plt.show()
    cv2.namedWindow("original",cv2.WINDOW_NORMAL)
    cv2.imshow("original", input_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()    
```

```python
def horizontally(input_image):
    flipped = cv2.flip(input_image, 1)
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(flipped); plt.title("horizontal flip");
    plt.show()
    cv2.namedWindow("horizontal flip",cv2.WINDOW_NORMAL)
    cv2.imshow("horizontal flip", flipped)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
horizontally(img)
```

```python
def transpose(input_image):
    transposed = cv2.transpose(input_image)
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(transposed); plt.title("transpose");
    plt.show()
    cv2.namedWindow("transpose",cv2.WINDOW_NORMAL)
    cv2.imshow("transpose", transposed)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def crop(input_image):
    width_scaled, height_scaled = input_image.shape[1] , input_image.shape[0]
    left_x, right_x = (width_scaled / 4), width_scaled - (width_scaled / 4) 
    top_y, bottom_y = (height_scaled / 4), height_scaled - (height_scaled / 4) 
    cropped = input_image[int(top_y):int(bottom_y), int(left_x):int(right_x)]
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(cropped); plt.title("crop");
    plt.show()
    cv2.namedWindow("crop",cv2.WINDOW_NORMAL)
    cv2.imshow("crop", cropped)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
crop(img)
```

```python
def right_shift(input_image):
    (rows, cols) = input_image.shape[:2]
    M = np.float32([[1, 0, 2], [0, 1, 0]])
    N = np.float32([[1, 0, 2-cols] , [0, 1, 0]])
    Right_shift = cv2.warpAffine(input_image, M, (cols, rows)) + cv2.warpAffine(input_image, N, (cols, rows))
    cv2.imwrite('Right_shift.tif', Right_shift)
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(Right_shift); plt.title("Right shift");
    plt.show()
    cv2.namedWindow("Right_shift",cv2.WINDOW_NORMAL)
    cv2.imshow("Right_shift", Right_shift)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
right_shift(img)
```

```python
def left_shift(input_image):
    (rows, cols) = input_image.shape[:2]
    M = np.float32([[1, 0, -2], [0, 1, 0]])
    N = np.float32([[1, 0, cols-2] , [0, 1, 0]])
    Left_shift = cv2.warpAffine(input_image, M, (cols, rows)) + cv2.warpAffine(input_image, N, (cols, rows))
    cv2.imwrite('Left_shift.tif', Left_shift)
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(Left_shift); plt.title("Left shift");
    plt.show()
    cv2.namedWindow("Left_shift",cv2.WINDOW_NORMAL)
    cv2.imshow("Left_shift", Left_shift)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
left_shift(img)
```

```python
def up_shift(input_image):
    (rows, cols) = input_image.shape[:2]
    M = np.float32([[1, 0, 0], [0, 1, -2]])
    N = np.float32([[1, 0, 0] , [0, 1, rows-2]])
    Up_shift = cv2.warpAffine(input_image, M, (cols, rows)) + cv2.warpAffine(input_image, N, (cols, rows))
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(Up_shift); plt.title("Up shift");
    plt.show()
    cv2.namedWindow("Up_shift",cv2.WINDOW_NORMAL)
    cv2.imshow("Up_shift", Up_shift)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def down_shift(input_image):
    (rows, cols) = input_image.shape[:2]
    M = np.float32([[1, 0, 0], [0, 1, 2]])
    N = np.float32([[1, 0, 0] , [0, 1, 2-rows]])
    down_shift = cv2.warpAffine(input_image, M, (cols, rows)) + cv2.warpAffine(input_image, N, (cols, rows))
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(down_shift); plt.title("Down shift");
    plt.show()
    cv2.namedWindow("Down_shift",cv2.WINDOW_NORMAL)
    cv2.imshow("Down_shift", down_shift)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def HW1_MSE(imageA, imageB):
    err = np.sum(((imageA - imageB) ** 2))
    err /= float(imageA.shape[0] * imageA.shape[1])
    print("MSE = %.2f" % (err))
    return err
```

```python
img1=cv2.imread(r"C:\Users\ALMAS\Documents\Untitled Folder\Hi.tif",1)
img2=cv2.imread(r"C:\Users\ALMAS\Documents\Untitled Folder\hybrid_image.tif",1)
m = HW1_MSE(img1,img2)
fig = plt.figure("MSE")
plt.suptitle("MSE: %.2f" % (m))
# show first image
ax = fig.add_subplot(1, 2, 1)
plt.imshow(img1, cmap = plt.cm.gray)
plt.axis("off")
# show the second image
ax = fig.add_subplot(1, 2, 2)
plt.imshow(img2, cmap = plt.cm.gray)
plt.axis("off")
# show the images
plt.show()
```

```python
def change_brightness(hi):
    matrix = np.ones(hi.shape, dtype= "uint8") * 20
    width_scaled, height_scaled = hi.shape[1] , hi.shape[0]
    darker = cv2.subtract(hi, matrix) 
    m = HW1_MSE(hi,darker)
    fig = plt.figure("MSE")
    #plt.suptitle("MSE: %.2f" % (m))
    # show first image
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(hi); plt.title("original");
    plt.subplot(132); plt.imshow(darker); plt.title("Darker");
    # show the second image
    
    # show the images
    
    plt.show()
   
```

```python
def get_avg(img, img1, img2):
    image_data = []
    image_data.append(img)
    image_data.append(img1)
    image_data.append(img2)
    avg_image = image_data[0]
    for i in range(len(image_data)):
        if i == 0:
            pass
        else:
            alpha = 1.0/(i + 1)
            beta = 1.0 - alpha
            avg_image = cv2.addWeighted(image_data[i], alpha, avg_image, beta, 0.0)
    cv2.imwrite('avg_image.tif', avg_image)
    #avg_image = cv2.imread(r"C:\Users\ALMAS\Documents\Untitled Folder\avg_image.tif",1)
    m = HW1_MSE(img,avg_image)
    fig = plt.figure("MSE")
    plt.suptitle("MSE: %.2f" % (m))
# show first image
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(avg_image); plt.title("Avg_image");
    plt.subplot(132); plt.imshow(img); plt.title("original");
    
    # show the second image
    
    # show the images
    
    plt.show()
   
```

```python
imgA=cv2.imread(r"C:\Users\ALMAS\Documents\Untitled Folder\Hi.tif",1)
imgB=cv2.imread(r"C:\Users\ALMAS\Documents\Untitled Folder\Right_shift.tif",1)
imgC=cv2.imread(r"C:\Users\ALMAS\Documents\Untitled Folder\Left_shift.tif",1)
get_avg(imgA, imgB, imgC)
```

```python
def resize(input_image):
    new_img = cv2.resize(input_image,None, fx=0.8,fy=0.8)
    cv2.imwrite('Resized_image.tif', new_img)
    cv2.namedWindow("resized_0.8",cv2.WINDOW_NORMAL)
    cv2.imshow("resized_0.8", new_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    #nearest
    nearest = cv2.resize(new_img, dsize=(450, 450), interpolation=cv2.INTER_NEAREST)
    cv2.imwrite('Nearest.tif', nearest)
    cv2.namedWindow("Nearest",cv2.WINDOW_NORMAL)
    cv2.imshow("Nearest", nearest)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    print("Nearest:")
    m = HW1_MSE(input_image,nearest)
    #bilinear
    bilinear = cv2.resize(new_img, dsize=(450, 450), interpolation=cv2.INTER_LINEAR)
    cv2.imwrite('Bilinear.tif', bilinear)
    cv2.namedWindow("Bilinear",cv2.WINDOW_NORMAL)
    cv2.imshow("Bilinear", bilinear)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    print("Bilinear:")
    n = HW1_MSE(input_image,bilinear)
    #bicubic
    bicubic = cv2.resize(new_img, dsize=(450, 450), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite('Bicubic.tif', bicubic)
    cv2.namedWindow("Bicubic",cv2.WINDOW_NORMAL)
    cv2.imshow("Bicubic", bicubic)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    print("Bicubic:")
    p = HW1_MSE(input_image,bicubic)
# show first image
    plt.figure(figsize=[18,5])
    plt.subplot(131); plt.imshow(nearest); plt.title("Nearest\n MSE: %.2f" % (m));
    plt.subplot(132); plt.imshow(bilinear); plt.title("Bilinear\n MSE: %.2f" % (n));
    plt.subplot(133); plt.imshow(bicubic); plt.title("Bicubic\n MSE: %.2f" % (p));
    
    # show the second image
    
    # show the images
    
    plt.show()
```
